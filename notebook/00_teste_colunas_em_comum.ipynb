```python
import pandas as pd
import numpy as np
from pathlib import Path
import sys

sys.path.append('../src')
import data_cleaner as dc
```

```python
RAW_DATA_DIR, PROCESSED_DATA_DIR = dc.caminho_para_dataset()
```

```python
try:
  df_2024_kaggle = pd.read_csv (PROCESSED_DATA_DIR / 'state_of_data_2024_rico_limpo.csv')
  df_2023_kaggle = pd.read_csv (PROCESSED_DATA_DIR / 'state_of_data_2023_rico_limpo.csv')
  df_2022_kaggle = pd.read_csv (PROCESSED_DATA_DIR / 'state_of_data_2022_rico_limpo.csv')
  df_2021_kaggle = pd.read_csv (PROCESSED_DATA_DIR / 'state_of_data_2021_rico_limpo.csv')
  print('arquivos encontrados')
except FileNotFoundError as e: 
  print("Arquivo nao encontrado " + e)
```

```python
colunas_2024 = set(df_2024_kaggle)
colunas_2023 = set(df_2023_kaggle)
colunas_2022 = set(df_2022_kaggle)
colunas_2021 = set(df_2021_kaggle)

colunas_comum = colunas_2024.intersection(colunas_2023, colunas_2022, colunas_2021)
```

```python
print(colunas_comum)
```

```python
len(colunas_comum)
```

```python
colunas_so_em_2024 = colunas_2024 - colunas_2023 - colunas_2022 - colunas_2021
colunas_so_em_2023 = colunas_2023 - colunas_2024 - colunas_2022 - colunas_2021
colunas_so_em_2022 = colunas_2022 - colunas_2023 - colunas_2024 - colunas_2021
colunas_so_em_2021 = colunas_2021 - colunas_2022 - colunas_2023 - colunas_2024
```

```python
print(f"Total de colunas em 2024: {len(colunas_2024)}")
print(f"Total de colunas em 2023: {len(colunas_2023)}")
print(f"Total de colunas em 2022: {len(colunas_2022)}")
print(f"Total de colunas em 2021: {len(colunas_2021)}")

print("-" * 50)
print(f"Colunas em comum: {len(colunas_comum)}")
print(f"Colunas exclusivas de 2024: {len(colunas_so_em_2024)}")
print(f"Colunas exclusivas de 2023: {len(colunas_so_em_2023)}")
print(f"Colunas exclusivas de 2022: {len(colunas_so_em_2022)}")
print(f"Colunas exclusivas de 2021: {len(colunas_so_em_2021)}")

print("-" * 50)
```

```python
print("Exclusivas de 2024:", sorted(list(colunas_so_em_2024)))
print("Exclusivas de 2023:", sorted(list(colunas_so_em_2023)))
print("Exclusivas de 2022:", sorted(list(colunas_so_em_2022)))
print("Exclusivas de 2021:", sorted(list(colunas_so_em_2021)))
```

```python
df_unificado = pd.concat([df_2024_kaggle, df_2023_kaggle, df_2022_kaggle, df_2021_kaggle], ignore_index=True, sort=False)
```

```python
print("União concluída com sucesso!")
print(f"Dimensões do DataFrame unificado: {df_unificado.shape}")
print('\nTeste de coluna')
print(df_unificado['ano'].value_counts())
```
